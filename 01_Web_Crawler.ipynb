Web Crawler

```python
import HTMLParser
import requests
from bs4 import BeautifulSoup
mat = []
N = -1

#Fiest layer
re1=requests.get('http://gametsg.techbang.com/ro/index.php?view=npc&k1=monster&race=Brute#detail')
re1.encoding = "utf-8"
soup=BeautifulSoup(re1.text, "html.parser")

#Find URL (Second layer)
bid_table = soup.findAll('a',{'class':'color18 mid1'})
for tag in bid_table:
    links = str(['http://gametsg.techbang.com/ro/'+tag['href'].encode('utf-8')])[2:-2]

#Second layer
    re2=requests.get(links)
    re2.encoding = "utf-8"
    soup=BeautifulSoup(re2.text, "html.parser")
    
#Name
    for t in soup.select('.mid1.color8'):
        mat.append([t.text.encode('utf-8').strip()])                    
        N+=1

#Drop&Map
    for line in soup.select('.gray_bottom2.h18'):
        for l in line.select('.color8'):
            
            mat[N].append(l.text.encode('utf-8').strip())                    
                

    #print str(mat).decode('string_escape')
```

```python
print str(mat).decode('string_escape')
f = open('Brute_list.txt','w')
f.write(str(mat).decode('string_escape'))
f.close()
```
